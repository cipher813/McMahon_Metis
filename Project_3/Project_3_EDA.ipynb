```python
# Tor servers
ts = pd.read_pickle('data/torservers.pkl')
print(ts.shape)
ts.head()
```

```python
flag_list = ['Exit','Fast','Guard','HSDir','Stable']
flag_dict = {}

for flag in flag_list:
    flag_dict[flag] = ts.loc[flag,:]
    flag_dict[flag]['date'] = flag_dict[flag].index
```

```python
for k, v in flag_dict.items():
    y = flag_dict[k]['relays']
    x = flag_dict[k]['date']

    fig, ax = plt.subplots(figsize=(20, 2))
    plt.plot(x,y)
    plt.title(k,fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Servers', fontsize=12)
    plt.savefig('charts/' + str(flag) + '.png')
    print(plt.show()) 
```

```python
ts = ts.groupby(['flag']).sum()
ts
```
