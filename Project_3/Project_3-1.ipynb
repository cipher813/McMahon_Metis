```python
td = pd.read_csv('data/tordata20180205.csv')
td['date'] = pd.to_datetime(td['date'])
start_date = pd.to_datetime('2017-01-01')
td = td[td['date'] >= start_date]
print(td.shape)
print(td.info())
print(td.columns)
td.head()
```

```python
td = td.groupby(["country","date"]).sum()
td = pd.DataFrame(td)
print(td.shape)
td.head()
```

```python

country_list_s = ['cn','ru','ir','sa','tr','in','id','gr','sd','be','cy','kr','de','us']
country_list_l = ['China','Russia','Iran','Saudi Arabia','Turkey','India','Indonesia','Greece','sd','Belgium','Cyprus','Korea','Germany','United States']
country_dict = {}
for country in country_list_s:
    country_dict[country] = td.loc[country,:]
    country_dict[country]['date'] = country_dict[country].index
```

```python
i = 0
for k, v in country_dict.items():
    y = country_dict[k]['clients']
    x = country_dict[k]['date']

    plt.plot(x,y)
    plt.title(k,fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Clients', fontsize=12)
    plt.tight_layout()
    plt.savefig('charts/' + k + str(start_date).replace('-','').split(' ')[0] + '.png')
    print(plt.show())
    i += 1
    
```
