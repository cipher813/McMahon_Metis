## Load Data

### Tor Metrics

```python
start_date = pd.to_datetime('2017-01-01') # remove earlier data due to noise
```

```python
# Tor Metrics Combined (Bridges?)
tb = pd.read_csv('https://metrics.torproject.org/stats/userstats-combined.csv') # tor metrics: clients
```

```python
# tb.head()
tb.describe()
```

```python
tb['date'] = pd.to_datetime(tb['date'])
tb = tb[tb['date'] >= start_date] # filter from start date to current
tb = tb.groupby(["country","node","date"]).sum()
tb = pd.DataFrame(tb)
tb.to_pickle('data/torcombined.pkl')
print(tb.shape)
# td.info()
tb.head()
```

```python
# Tor Metrics: Clients
td = pd.read_csv('https://metrics.torproject.org/stats/clients.csv') # tor metrics: clients
```

```python
td.head()
td.describe()
```

```python
td['date'] = pd.to_datetime(td['date'])
# start_date = pd.to_datetime('2017-01-01') # remove earlier data due to noise
td = td[td['date'] >= start_date] # filter from start date to current
td = td.groupby(["country","node","date"]).sum()
td = pd.DataFrame(td)
td.to_pickle('data/torclients.pkl')
print(td.shape)
# td.info()
td.head()
```

```python
# Tor Metrics: Servers
ts = pd.read_csv('https://metrics.torproject.org/stats/servers.csv') # tor metrics: servers
```

```python
# ts.head()
ts.describe()
```

```python
ts['date'] = pd.to_datetime(ts['date'])
ts = ts[ts['date'] >= start_date] # filter from start date to current
ts = ts.groupby(["flag","date"]).sum()
ts = pd.DataFrame(ts)
ts.to_pickle('data/torservers.pkl')
print(ts.shape)
# ts.info()
# ts.head()
ts.describe()
```
